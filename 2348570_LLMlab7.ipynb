```python
from datasets import load_dataset
from transformers import BertTokenizer
```

```python
dataset = load_dataset('imdb')
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def tokenize_and_encode(example):
    # Tokenize and encode the text
    return tokenizer(example['text'], padding='max_length', truncation=True)

# Apply the function to the 'train' split of the dataset
encoded_dataset = dataset['train'].map(tokenize_and_encode, batched=True)
```

```python
print(encoded_dataset[0])
```

```python
# Function to tokenize the text
def tokenize_text(example):
    # Tokenize the text (convert to list of tokens)
    tokens = tokenizer.tokenize(example['text'])

    # Add tokens to the output for inspection
    return {'tokens': tokens}

# Apply the function to the 'train' split (without batching)
tokenized_dataset = dataset['train'].map(tokenize_text)

# Inspect the first example with tokens
print(tokenized_dataset[0]['tokens'])
```
